```python
import numpy as np
import matplotlib.pyplot as plt

import h5py
```

```python
file = h5py.File('prior_pos_100.h5','r')
prior_sd_mu_P_100 = np.array(file['prior_sd_mu_P'])
pos_sd_mu_P_100 = np.array(file['pos_sd_mu_P'])
prior_sd_h_100 = np.array(file['prior_sd_h'])
pos_sd_h_100 = np.array(file['pos_sd_h'])
file.close()
```

```python
file = h5py.File('prior_pos_500.h5','r')
prior_sd_mu_P_500 = np.array(file['prior_sd_mu_P'])
pos_sd_mu_P_500 = np.array(file['pos_sd_mu_P'])
prior_sd_h_500 = np.array(file['prior_sd_h'])
pos_sd_h_500 = np.array(file['pos_sd_h'])
file.close()
```

```python
file = h5py.File('prior_pos_1000.h5','r')
prior_sd_mu_P_1000 = np.array(file['prior_sd_mu_P'])
pos_sd_mu_P_1000 = np.array(file['pos_sd_mu_P'])
prior_sd_h_1000 = np.array(file['prior_sd_h'])
pos_sd_h_1000 = np.array(file['pos_sd_h'])
file.close()
```

```python
plt.scatter(prior_sd_mu_P_100, pos_sd_mu_P_100, label='100 Events')
plt.scatter(prior_sd_mu_P_500, pos_sd_mu_P_500, label='500 Events')
plt.scatter(prior_sd_mu_P_1000, pos_sd_mu_P_1000, label='1000 Events')
plt.xlabel('prior_sd_mu_P')
plt.ylabel('pos_sd_mu_P')
plt.title('prior_sd_h = 0.5')
plt.legend()
plt.savefig('mu_P.pdf')
```

```python
plt.scatter(prior_sd_h_100, pos_sd_h_100, label='100 Events')
plt.scatter(prior_sd_h_500, pos_sd_h_500, label='500 Events')
plt.scatter(prior_sd_h_1000, pos_sd_h_1000, label='1000 Events')
plt.xlabel('prior_sd_h')
plt.ylabel('pos_sd_h')
plt.title('prior_sd_mu_P = 0.5')
plt.legend()
plt.savefig('h.pdf')
```
